# Styles and Colormaps

## colormaps

```python
# !pip install gapminder
from gapminder import gapminder
gapminder.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
gapminder['norm_pop'] = gapminder['pop'].apply(lambda x: x*2/1000000)
gm2007=gapminder[gapminder.year==2007]
```

```python
fig = plt.figure(figsize=(10,6));
ax = plt.gca()
gm2007.plot.scatter('gdpPercap', 'lifeExp', c='pop', 
                    colormap='viridis', ax=ax)
ax.set_xlabel('GDP per capita')
ax.set_ylabel('Life Expectancy')
ax.set_xscale('log');
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,6));
gm2007.plot.scatter('gdpPercap', 'lifeExp', 
                    c='norm_pop', s='norm_pop',
                    colormap='inferno', 
                    alpha=0.75, ax=ax)
ax.set_xlabel('GDP per capita')
ax.set_ylabel('Life Expectancy')
ax.set_xscale('log');
```

More [colormaps here](https://matplotlib.org/2.0.2/examples/color/colormaps_reference.html).
